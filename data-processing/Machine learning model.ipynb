```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, mean_absolute_error 
from sklearn import preprocessing 
```

```python
df = pd.read_csv('sunflower_data.csv') 
```

```python
df_data_only = df.drop(['centroid_x','centroid_y','crop_type'], axis=1)
df_data_only
```

```python
X = df_data_only.drop('area', axis=1)
y = df_data_only['area']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
    X, y, test_size=0.2, random_state=10) 
```

```python
model = LinearRegression() 
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test) 
```

```python
np.var(predictions)
```

```python
print( 
  'mean_squared_error : ', mean_squared_error(y_test, predictions)) 
print( 
  'mean_absolute_error : ', mean_absolute_error(y_test, predictions)) 
```

```python
def add_gradient_circle(ax, center, max_radius, color, num_steps=10):
    """
    Adds a gradient circle to the axis.

    :param ax: The axis to add the circle to.
    :param center: A tuple (longitude, latitude) for the center of the circle.
    :param max_radius: The maximum radius of the gradient circle.
    :param color: The color of the circle.
    :param num_steps: The number of concentric circles to draw.
    """
    for i in range(num_steps, 0, -1):
        radius = max_radius * (i / num_steps)
        alpha = 1 - (i / num_steps)
        circle = plt.Circle(center, radius, color=color, transform=ccrs.Geodetic(), alpha=alpha, linewidth=0)
        ax.add_patch(circle)
```

```python
import math
def radius_from_area(area):
     # Average radius of the Earth in meters
    earth_radius = 6371000

    # Calculate the radius in meters
    radius_meters = math.sqrt(area / math.pi)

    # Convert the radius to degrees at the equator
    radius_degrees = (radius_meters / (2 * math.pi * earth_radius)) * 360
    return radius_degrees
```

```python

import cartopy.crs as ccrs

# Set up the map
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

ax.coastlines()

# Plot each point
for idx, row in df.head(10).iterrows():
    color = "red"  # Default color is black
    area = row['SVR_predicted_area']
    radius = radius_from_area(area)*5
    # print(radius)
    add_gradient_circle(ax, (row['centroid_y'], row['centroid_x']+0.01), radius, color)

plt.show()
```

```python
from pytorch_tabnet.tab_model import TabNetClassifier, TabNetRegressor
```

```python
# Initialize the TabNetRegressor model
model = TabNetRegressor()

y_train_reshaped = y_train.values.reshape(-1, 1)
y_test_reshaped = y_test.values.reshape(-1, 1)

# Train the model on the training data
model.fit(X_train.values, y_train_reshaped, eval_set=[(X_test.values, y_test_reshaped)], patience=100, max_epochs=100)

```

```python
# Make predictions on the test set
y_pred = model.predict(X_test.values)

```

```python
np.var(y_pred)
```

```python
# Evaluate the model's performance (you can use any regression metric)
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
print('Mean_absolute_error : ', mean_absolute_error(y_test, y_pred)) 
```
